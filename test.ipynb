```
def display_multiplication_table(number, limit=10):
    print(f"Multiplication Table for {number}:")
    for i in range(1, limit + 1):
        print(f"{number} x {i} = {number * i}")

# Example usage:
num_to_display = 7
display_multiplication_table(num_to_display)
```

## Simple Expert System for Disease Diagnosis

This code implements a basic expert system that attempts to diagnose a few common ailments (FLU, COMMON COLD, VIRAL FEVER) based on a set of 'if-then' rules and user-provided symptoms. The system will prompt you for symptoms like fever, cough, headache, and sneezing.

```
def expert_system():
    print("=== Medical Diagnosis Expert System ===")
    fever = input("Do you have fever? (yes/no): ").lower()
    cough = input("Do you have cough? (yes/no): ").lower()
    headache = input("Do you have headache? (yes/no): ").lower()
    sneezing = input("Do you have sneezing? (yes/no): ").lower()

    # Inference Engine (Rules)
    if fever == "yes" and cough == "yes" and headache == "yes":
        diagnosis = "You may have FLU."
    elif sneezing == "yes" and cough == "yes" and fever == "no":
        diagnosis = "You may have COMMON COLD."
    elif fever == "yes" and headache == "yes":
        diagnosis = "You may have VIRAL FEVER."
    else:
        diagnosis = "Condition unclear. Please consult a doctor."

    print("\nDiagnosis Result:")
    print(diagnosis)

# Run the expert system
expert_system()
```

```
def is_palindrome(number):
    # Convert the number to a string to easily reverse it
    num_str = str(number)
    # Compare the string with its reverse
    return num_str == num_str[::-1]

# Example usage:
num1 = 121
if is_palindrome(num1):
    print(f"The number {num1} is a palindrome.")
else:
    print(f"The number {num1} is not a palindrome.")

num2 = 12345
if is_palindrome(num2):
    print(f"The number {num2} is a palindrome.")
else:
    print(f"The number {num2} is not a palindrome.")

num3 = 1221
if is_palindrome(num3):
    print(f"The number {num3} is a palindrome.")
else:
    print(f"The number {num3} is not a palindrome.")
```

```
def check_odd_even(number):
    if number % 2 == 0:
        return "Even"
    else:
        return "Odd"

# Example usage:
num_to_check = 15
status = check_odd_even(num_to_check)
print(f"The number {num_to_check} is {status}.")

num_to_check = 20
status = check_odd_even(num_to_check)
print(f"The number {num_to_check} is {status}.")
```

```
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n-1)

# Example usage:
number = 5
result = factorial(number)
print(f"The factorial of {number} is {result}")
```
